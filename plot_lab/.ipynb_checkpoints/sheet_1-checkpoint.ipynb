```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import struct
from pathlib import Path
import numpy as np
```

```python
data = Path("data.bin").read_bytes()
data_Ns =  Path("data_Ns.bin").read_bytes()
```

```python
N = int.from_bytes(data[:4], byteorder='little', signed=False)
M = int.from_bytes(data[4:8], byteorder='little', signed=False)
data_array = np.frombuffer(data[8:], dtype=np.float64)
data_Ns_array = np.frombuffer(data_Ns, dtype=np.int32)
data_array = data_array.reshape(N,M)
```

```python
print(data_Ns_array.shape)
print(data_array[:,0].shape)
```

```python
plt.plot(data_Ns_array, data_array, linestyle="", marker="o", markersize=1)
plt.xscale("log")
plt.ylabel(r"$f_{mean}(N)$")
plt.xlabel("N")
plt.savefig("scatter.png")
```

```python
def fit(N, a, b):
    return a * N**b
```

```python
popt, pcov = curve_fit(fit, data_Ns_array, np.std(data_array, axis=1), sigma=np.std(data_array, axis=1))
```

```python
plt.plot(data_Ns_array, np.std(data_array, axis=1), linestyle="", marker="o", markersize=1)
plt.plot(data_Ns_array, fit(data_Ns_array, *popt), linestyle="-", marker="", markersize=1)
plt.xscale("log")
plt.yscale("log")
plt.ylabel(r"$s_{f_{mean}}$")
plt.xlabel("N")
plt.title(rf"a = {popt[0]:.3f}, b = {popt[1]:.3f} ")
plt.savefig("fit.png")
```

$s_{f_{mean}} \propto a N^b \approx \frac{1}{2} N^{-0.5}$
